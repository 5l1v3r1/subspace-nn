```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable

from torchvision import datasets, models, transforms, utils
import numpy as np
import matplotlib.pyplot as plt
import os

from utils import *

reseed = lambda: np.random.seed(seed=123) ; ms = torch.manual_seed(123) # for reproducibility
reseed()
```

## Hyperparameters

This gets a little hacky because I'm trying to mimic `argparse`, which I use in `main.py` to set hyperparameters. Doing this is worthwhile because it allows me to keep the rest of the syntax identical between this notebook and the `main.py` file.

```python
class DummyArgparse():
    def __init__(self):
        pass
args = DummyArgparse()

args.fig_dir = 'figures/'      # directory to save figures
args.batch_size = 256          # batch size for training
args.omega_dims = [None, 1000, 3000] # sizes for omega subspaces
args.lr = 3e-3                 # learning rate
args.test_every = 100          # record test accuracy after this number of steps
args.epochs = 10               # number of epochs to train

# book-keeping
os.makedirs(args.fig_dir) if not os.path.exists(args.fig_dir) else None
args.input_dim = 28**2
args.target_dim = 10
global_step = 0
total_steps = int(60000*args.epochs/args.batch_size)
print("using {} epochs ({} steps)".format(args.epochs, total_steps))
```

## Model zoo

```python
class fcNN(nn.Module):
    def __init__(self, h_dim, batch_size):
        super(fcNN, self).__init__()
        self.batch_size = batch_size
        self.fc1 = nn.Linear(784, h_dim)
        self.fc2 = nn.Linear(h_dim, 10)
        
    def param_count(self):
        return sum([np.prod(p.size()) for p in self.parameters()])

    def forward(self, X):
        h1 = F.relu(self.fc1(X))
        h2 = F.log_softmax(self.fc2(h1))
        return h2
    
class Conv4NN(nn.Module):
    def __init__(self, batch_size):
        super(Conv4NN, self).__init__()
        self.batch_size = batch_size
        self.conv1 = nn.Conv2d(1, 8, kernel_size=4)
        self.conv2 = nn.Conv2d(8, 16, kernel_size=4)
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(16*4*4, 32)
        self.fc2 = nn.Linear(32, 10)

    def param_count(self):
        return sum([np.prod(p.size()) for p in self.parameters()])

    def forward(self, X):
        X = X.resize(self.batch_size, 1, 28, 28)
        z1 = F.max_pool2d(self.conv1(X), 2) ; h1 = F.relu(z1)
        z2 = F.max_pool2d(self.conv2_drop(self.conv2(h1)), 2)
        h2 = F.relu(z2).view(-1, 16*4*4)
        h3 = F.relu(self.fc1(h2)) ; h3 = F.dropout(h3, training=self.training)
        h4 = F.log_softmax(self.fc2(h3))
        return h4
    
class Conv3NN(nn.Module):
    def __init__(self, batch_size):
        super(Conv3NN, self).__init__()
        self.batch_size = batch_size
        self.conv1 = nn.Conv2d(1, 6, kernel_size=4)
        self.conv2 = nn.Conv2d(6, 6, kernel_size=4)
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(6*4*4, 25)
        self.fc2 = nn.Linear(25, 10)

    def param_count(self):
        return sum([np.prod(p.size()) for p in self.parameters()])

    def forward(self, X):
        X = X.resize(self.batch_size, 1, 28, 28)
        z1 = F.max_pool2d(self.conv1(X), 2) ; h1 = F.relu(z1)
        z2 = F.max_pool2d(self.conv2_drop(self.conv2(h1)), 2)
        h2 = F.relu(z2).view(-1, 6*4*4)
        h3 = F.relu(self.fc1(h2)) ; h3 = F.dropout(h3, training=self.training)
        h4 = F.log_softmax(self.fc2(h3))
        return h4
```

## Train several models
Train several small MNIST models. Plot their loss and accuracies and track their convergence properties.

```python
# train models with different omega subspace sizes, saving loss and accuracies for each
fc2 = fcNN(32, batch_size=args.batch_size)
fc3 = fcNN(16, batch_size=args.batch_size)
conv3 = Conv3NN(batch_size=args.batch_size)
conv4 = Conv4NN(batch_size=args.batch_size)
models = [fc2, fc3, conv3, conv4]

print('# parameters: ', [m.param_count() for m in models])
histories = []
for model in models:
    reseed()
    optimizer = torch.optim.Adam(model.parameters(), lr=args.lr, betas=(0.9, 0.99), weight_decay=0)
    h = train(dataloader, model, optimizer, global_step, total_steps, args.test_every)
    histories.append(h)
    print('\n\n')
```

## Compare losses

```python
f = plt.figure(figsize=[8,5])
for i, model in enumerate(models):
    loss_hist = histories[i][0]
    xy = np.stack(loss_hist)
    plt.plot(xy[:,0], xy[:,1], linewidth=3.0, label='[$\Theta$]={}'.format(model.param_count()))

title = "MNIST zoo training losses"
plt.title(title, fontsize=16)
plt.xlabel('train step', fontsize=14) ; plt.setp(plt.gca().axes.get_xticklabels(), fontsize=14)
plt.ylabel('loss', fontsize=14) ; plt.setp(plt.gca().axes.get_yticklabels(), fontsize=14)
plt.ylim([0,2.5])
plt.legend()

plt.show()
```

## Compare accuracies

```python
f = plt.figure(figsize=[8,5])
for i, model in enumerate(models):
    acc_hist = histories[i][1]
    xy = np.stack(acc_hist)
    plt.plot(xy[:,0], xy[:,1], linewidth=3.0, label='[$\Theta$]={}'.format(model.param_count()))

title = "MNIST zoo test accuracies"
plt.title(title, fontsize=16)
plt.xlabel('train step', fontsize=14) ; plt.setp(plt.gca().axes.get_xticklabels(), fontsize=14)
plt.ylabel('accuracy (%)', fontsize=14) ; plt.setp(plt.gca().axes.get_yticklabels(), fontsize=14)

plt.ylim([0,100])
plt.legend()

plt.show()
```
